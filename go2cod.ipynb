# DATA VISUALIZATION

## Avg. Session Length

```python
hist = data['Avg. Session Length'].hist(edgecolor='black')


bin_edges = hist.patches
x_ticks = []
for patch in bin_edges:
    bin_range = patch.get_bbox().bounds  
    lower_bound = int(bin_range[0])      
    upper_bound = int(bin_range[0] + patch.get_width())  
    x_ticks.append(f'{lower_bound} - {upper_bound}')  

plt.xticks(ticks=[patch.get_x() + patch.get_width() / 2 for patch in bin_edges], labels=x_ticks, rotation=45, ha='right')

plt.title('Distribution of Avg. Session Length')
plt.xlabel('Avg. Session Length')
plt.ylabel('Frequency')
for p in hist.patches:
    height = p.get_height()
    hist.annotate(f'{int(height)}', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom')
plt.show()
```

```python
sns.boxplot(data['Avg. Session Length'])
plt.title('Boxplot of Avg. Session Length')
plt.show()
```

# Correlation Analysis

```python
# Correlation Matrix
correlation_matrix = data[['Avg. Session Length', 'Time on App', 'Time on Website', 'Length of Membership', 'Yearly Amount Spent']].corr()


# Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=1)
plt.title("Correlation Heatmap for All Variables", fontsize=16)
plt.show()
```
